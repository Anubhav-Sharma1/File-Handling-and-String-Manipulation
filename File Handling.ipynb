```python
import pandas as pd
import string
import re
from collections import defaultdict
```

```python
conversation = pd.read_csv("C:/Users/Dell/OneDrive/Desktop/conv.txt",delimiter =':',header=None)
```

```python
conversation
```

```python
conversation.columns=['Speaker','Dialogue']
conversation.head()
```

```python
conversation['Speaker'].value_counts()
```

```python
#Task1 to find total number of unique speakers in a GOT script present in conv.txt
speaker_set = set(conversation['Speaker'])
count=0
for speaker in speaker_set:
        count=count+1
print("Total number of speakers : ",count)        
```

```python
dialogue_list = list(conversation['Dialogue'])
```

```python
dialogue_list_lower = []
dialogue_list_final = []

for lower_d in dialogue_list:
    str_lower = lower_d.lower()
    dialogue_list_lower.append(str_lower)
    
dialogue_list_lower_clean = (re.sub("[-()\"#/@;:<>{}`+=~|.!?,...]","",_) for _ in dialogue_list_lower)

for d in dialogue_list_lower_clean:
    str_test=('  '.join(dict.fromkeys(d.split())))
    dialogue_list_final.append(str_test)
```

```python
dialogue_list_final
```

```python
speaker_list = list(conversation['Speaker'])
speaker_list
```

```python
#combine speaker and their dialogues with the help of dictionary
speaker_dialogue_dic = defaultdict(list)
for key,value in zip(speaker_list,dialogue_list_final):
    speaker_dialogue_dic[key].append(value)
for key,values in speaker_dialogue_dic.items():
    for i in values:
        print(key+":"+i)
```

```python
#task 2 finding unique words from dialogues spoken by speakers
str_words = " "
list=[]
list2=[]
myfile = open("C:/Users/Dell/OneDrive/Desktop/WILL.txt",'w')
for key,values in speaker_dialogue_dic.items():
    if(key=="WILL"):
        for sentence in values:
            for word in range(len(sentence)):
                if(sentence[word]==" "):
                    list.append(str_words)
                    list2=set(list)
                    str_words = " "
                    
                else:
                    str_words = str_words + sentence[word]
for word in list2:
    print(word)
    myfile.write("%s\n"%word)
myfile.close()                    
```
